## KNN (K Nearest Neighbor)

1. Pengertian KNN (K Nearest Neighbor)

1.1 KNN merupakan sebuah metode pengelompokan terhadap sekumpulan data berdasarkan pembelajaran data yang sudah terklasifikasi sebelumnya

1.2 Dekat atau jauhnya tetangga biasanya dihitung berdasarkan Euclidean Distance.

1.3 Rumus Euclidean Distance adalah sebagai berikut:

$$D(x,y) = \sqrt(\Sigma_I^n(xi-yi)^2)$$


1.4 dimana D(a,b) adalah jarak skalar dari dua buah vektor data a dan b yang berupa matrik berukuran d dimensi

1.5 berikut adalah algoritma KNN:

- Menentukan parameter k (jumlah tetangga paling dekat)

- menghitung kuadrat jarak euclidean objek terhadap data training yang diberikan

- mengurutkan secara ascending

- mengumpulkan kategori Y (Klasifikasi nearest neighbor berdasarkan nilai k)

### Teori

Deteksi outlier adalah proses mengidentifikasi dan menangani data yang tidak biasa atau tidak terduga dalam sebuah kumpulan data.

rumus manhattan distance:

$$d(x,y) = \sqrt{\sum_{i=1}^{n}(x_i - y_i)}$$

### Perhitungan LOF menggunakan Scikit Learn

Berikut merupakan implementasi dari perhitungan LOF dengan menggunakan library Scikit Learn

```
from sklearn.neighbors import LocalOutlierFactor
X = [
    (63,	33,	6),
    (58,	27,	51),
    (71,	3,	59),
    (63,	29,	56),
    (65,	3,	58),
    (76,	3,	66),
    (49,	25,	45),
    (73,	29,	63),
    (67,	25,	58),
    (72,	36,	61)
]

lof = LocalOutlierFactor(n_neighbors=2, contamination=0.1)

y_pred = lof.fit_predict(X)

print("Predicted table:", y_pred)

print("Negative LOF scores:", -lof.negative_outlier_factor_)
```

```
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import csv

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Initialize the Naive Bayes classifier
nb_classifier = GaussianNB()

# Split the data into training and test sets (50% training, 50% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

# Train the classifier
nb_classifier.fit(X_train, y_train)

# Predict the classes for the training set
y_train_pred = nb_classifier.predict(X_train)

# Combine features and target into one list
train_data = np.column_stack((X_train, y_train))

# Write the training data to a CSV file
with open('training_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(iris.feature_names + ['Target'])
    # Write data
    writer.writerows(train_data)

# Calculate the accuracy of the classifier on the training set
training_accuracy = accuracy_score(y_train, y_train_pred)
print("Training Accuracy:", training_accuracy)

# Sample data for prediction
sample = np.array([[4.4, 2.9, 1.4, 0.2]])

# Predict the class for the sample
prediction = nb_classifier.predict(sample)
print('Prediction:', prediction)
```

```
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import csv

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Initialize the Naive Bayes classifier
nb_classifier = GaussianNB()

# Split the data into training and test sets (80% training, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the classifier
nb_classifier.fit(X_train, y_train)

# Predict the classes for the training set
y_test_pred = nb_classifier.predict(X_test)

# Combine features and target into one list
train_data = np.column_stack((X_train, y_train))

# Write the training data to a CSV file
with open('testing_data80.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(iris.feature_names + ['Target'])
    # Write data
    writer.writerows(train_data)

# Calculate the accuracy of the classifier on the training set
training_accuracy = accuracy_score(y_test, y_test_pred)
print("Training Accuracy:", training_accuracy)

# Sample data for prediction
sample = np.array([[4.4, 2.9, 1.4, 0.2]])

# Predict the class for the sample
prediction = nb_classifier.predict(sample)
print('Prediction:', prediction)
```
